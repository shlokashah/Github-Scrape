```python
contribution_year = []
contribution_year.append(temp.find(class_='js-year-link filter-item px-3 mb-2 py-2 selected')['href'])
contribution_year
```

```python
year = []
# print(temp)
year.append(temp.find_all(class_='js-year-link filter-item px-3 mb-2 py-2'))
contribution_year.append(year[0][0]['href'])
contribution_year.append(year[0][1]['href'])
print()
print(contribution_year)
no_of_years = len(contribution_year)
print(no_of_years)
```

```python
current = [2020]
flag = current[0]-1
for i in range(no_of_years-1):
    current.append(flag)   
    flag=flag-1
current
```

```python
contribution = []
for i in contribution_year:
    year_url = 'https://github.com'+i
    page_year = requests.get(year_url)
    soup_year = BeautifulSoup(page_year.content,'html.parser')
    contribution.append(soup_year.find(class_="f4 text-normal mb-2").text.strip()[:3])
contribution 
```

```python
fig = plt.figure()
plt.bar(current[::-1],contribution[::-1])
plt.xlabel("Year")
plt.xticks(current[::-1])
plt.ylabel("No of Commits")
plt.title("No of Commits each year")
plt.show()

```
